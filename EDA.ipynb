# EDA



```python
import warnings

warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from matplotlib.ticker import PercentFormatter
plt.rcParams.update({ "figure.figsize" : (8, 5),"axes.facecolor" : "white", "axes.edgecolor":  "black"})
plt.rcParams["figure.facecolor"]= "w"
pd.plotting.register_matplotlib_converters()
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv('data/eda.csv')
df.tail
df.id.count()
```

```python
df.columns
```

```python
plt.hist(df['bedrooms'], bins=30, alpha=0.6, color='g')
plt.title('Histogram of Bedrooms')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()

max_value = df['bedrooms'].max()
min_value = df['bedrooms'].min()

print(f"Maximum value in the column: {max_value}")
print(f"Minimum value in the column: {min_value}")

# Count occurrences of each number of bedrooms
bedroom_counts = df['bedrooms'].value_counts()
print("\nOccurrences of each number of bedrooms:")
print(bedroom_counts)
```

```python
plt.hist(df['bathrooms'], bins=30, alpha=0.6, color='g')
plt.title('Histogram of Bathrooms')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()

max_value = df['bathrooms'].max()
min_value = df['bathrooms'].min()

print(f"Maximum value in the column: {max_value}")
print(f"Minimum value in the column: {min_value}")

bathrooms = df['bathrooms'].value_counts()
print("\nOccurrences of each number of bathrooms:")
print(bathrooms)
```

```python
plt.hist(df['floors'], bins=30, alpha=0.6, color='g')
plt.title('Histogram of Floors')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()

max_value = df['floors'].max()
min_value = df['floors'].min()

print(f"Maximum value in the column: {max_value}")
print(f"Minimum value in the column: {min_value}")

floors = df['floors'].value_counts()
print("\nOccurrences of each number of floors:")
print(floors)
```

```python
waterfront_counts = df['waterfront'].value_counts()

labels = [f'{index}: {count}' for index, count in waterfront_counts.items()]

plt.figure(figsize=(4, 4))
plt.pie(waterfront_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgrey'])
plt.title('Distribution of Waterfront Properties')
plt.show()

waterfront = df['waterfront'].value_counts()
print("\nOccurrences of each number of waterfront:")
print(waterfront)
```

```python
condition = df['condition'].value_counts()

labels = [f'{index}: {count}' for index, count in condition.items()]

plt.figure(figsize=(5, 5))
patches, texts, autotexts = plt.pie(condition, autopct='%1.2f%%', startangle=90)

plt.title('Distribution of condition')
for autotext in autotexts:
    autotext.set_text('')

plt.legend(patches, labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

print("\nOccurrences of each number of condition:")
print(condition)
```

```python
df['grade_category'] = df['grade'].apply(lambda x: '7 or higher' if x >= 7 else 'Below 7')

grade_counts = df['grade_category'].value_counts()
grade_counts_sorted = grade_counts.sort_index()

labels = [f'{index}: {count}' for index, count in grade_counts_sorted.items()]

plt.figure(figsize=(5, 5))
patches, texts, autotexts = plt.pie(grade_counts_sorted, labels=labels, autopct='%1.2f%%', startangle=90)

plt.title('Distribution of grade categories')

plt.legend(patches, labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()

print("\nOccurrences of each grade category:")
print(grade_counts_sorted)
```

```python
df['renovated_category'] = df['yr_renovated'].apply(lambda x: 'Renovated' if x > 0 else 'Not Renovated')

grade_counts = df['renovated_category'].value_counts()
grade_counts_sorted = grade_counts.sort_index()

labels = [f'{index}: {count}' for index, count in grade_counts_sorted.items()]

plt.figure(figsize=(5, 5))
patches, texts, autotexts = plt.pie(grade_counts_sorted, labels=labels, autopct='%1.2f%%', startangle=90)

plt.title('Distribution of renovated categories')

plt.legend(patches, labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()

print("\nOccurrences of each renovated category:")
print(grade_counts_sorted)
```

```python
filtered_df = df[(df['waterfront'] > 0) & (df['yr_renovated'] > 0) & (df['grade'] > 6)]

num_filtered = len(filtered_df)
num_not_filtered = len(df) - num_filtered
labels = ['Filtered', 'Not Filtered']
sizes = [num_filtered, num_not_filtered]

plt.figure(figsize=(5, 5))

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d} ({p:.2f}%)'.format(p=pct, v=val)
    return my_format

patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct=autopct_format(sizes), startangle=90)

plt.title('Filtered vs Not Filtered Data')

plt.legend(patches, labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
avg_price_per_room = df.groupby('bedrooms')['price'].mean().reset_index()
avg_price_per_room_filtered = filtered_df.groupby('bedrooms')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))

plt.plot(avg_price_per_room['bedrooms'], avg_price_per_room['price'], label='All Houses', marker='o')
plt.plot(avg_price_per_room_filtered['bedrooms'], avg_price_per_room_filtered['price'], label='Target', marker='s')

plt.title('Average Price vs. Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Average Price')
plt.legend()
plt.show()
```

```python
bins = np.linspace(df['sqft_living'].min(), df['sqft_living'].max(), 101)  # 100 bins

df['sqft_bin'] = pd.cut(df['sqft_living'], bins)
filtered_df['sqft_bin'] = pd.cut(filtered_df['sqft_living'], bins)

avg_price_per_sqft = df.groupby('sqft_bin')['price'].mean().reset_index()
avg_price_per_sqft_filtered = filtered_df.groupby('sqft_bin')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))

bin_midpoints = [(bins[i] + bins[i+1])/2 for i in range(0, len(bins)-1, 10)]  # Midpoints for label bins
labels = [f"{int(bins[i])}-{int(bins[i+10])}" for i in range(0, len(bins)-1, 10)]  # Labels for every 10th bin

plt.plot(range(len(avg_price_per_sqft)), avg_price_per_sqft['price'], label='Unfiltered', marker='o')
plt.plot(range(len(avg_price_per_sqft_filtered)), avg_price_per_sqft_filtered['price'], label='Filtered', marker='s')

plt.title('Average Price vs. Square Footage')
plt.xlabel('Square Footage Bin Range')
plt.ylabel('Average Price')
plt.legend()

plt.xticks(ticks=np.arange(0, len(avg_price_per_sqft), 10), labels=labels, rotation=45)  # Apply custom ticks

plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
filtered_df['date'] = pd.to_datetime(filtered_df['date'])

df['month_year'] = df['date'].dt.to_period('M')
filtered_df['month_year'] = filtered_df['date'].dt.to_period('M')

avg_price_per_month_year = df.groupby('month_year')['price'].mean().reset_index()
avg_price_per_month_year_filtered = filtered_df.groupby('month_year')['price'].mean().reset_index()

avg_price_per_month_year['month_year'] = avg_price_per_month_year['month_year'].dt.to_timestamp()
avg_price_per_month_year_filtered['month_year'] = avg_price_per_month_year_filtered['month_year'].dt.to_timestamp()

plt.figure(figsize=(12, 6))

plt.plot(avg_price_per_month_year['month_year'], avg_price_per_month_year['price'], label='All', marker='o')
plt.plot(avg_price_per_month_year_filtered['month_year'], avg_price_per_month_year_filtered['price'], label='Filtered', marker='s')

plt.title('Average Price Over Time (Monthly)')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()

plt.grid(True, which="both", ls="--")

plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator(interval=1))
plt.gcf().autofmt_xdate()

plt.show()
```
